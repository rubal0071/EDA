### Road Accident Analysis in 2013

```python
top_n = 10
order = consolo_state_wise_acc.nlargest(top_n, 'State/UT-Wise Total Number of Road Accidents during - 2013')['States/Uts']
plt.figure(figsize=(15,5))
sns.barplot(x="States/Uts",y="State/UT-Wise Total Number of Road Accidents during - 2013", data=consolo_state_wise_acc,
           order=order)
plt.xlabel("States/ UTs")
plt.ylabel("Number of Accidents")
plt.title("State/ UT-wise Number of Accidents in 2013", fontsize=15, fontweight="bold")
plt.show()
```

```python
deaths_without_total=deaths_by_situation[deaths_by_situation["CrimeHead"]!="Total"]
deaths_crime_head_total=deaths_without_total[deaths_without_total["CrimeHead"].str.contains("Total")]
other_motor_veh=deaths_without_total[(deaths_without_total["CrimeHead"]=="Other Motor Vehicles") 
                                     |(deaths_without_total["CrimeHead"]=="Bicycle")
                                    |(deaths_without_total["CrimeHead"]=="Hand Drawn Vehicle/Cycle rickshaw (Passenger)")
                                    |(deaths_without_total["CrimeHead"]=="Animal Drawn Vehicle")
                                    |(deaths_without_total["CrimeHead"]=="Pedestrian")
                                    |(deaths_without_total["CrimeHead"]=="Others")]
```

```python
new_df=pd.concat([deaths_crime_head_total,other_motor_veh], axis=0)
new_df=new_df[(new_df["States/UTs"]!="Total (All India)") & (new_df["States/UTs"]!="Total (UTs)") & (new_df["States/UTs"]!="Total (States)")]
new_df=new_df[~new_df["CrimeHead"].str.contains("Sub Total")]
```

```python
death_veh_summ=new_df.groupby(["CrimeHead"])["Offenders (Driver/Pedestrian) Died_Total","Victims Died_Total"].sum()
death_veh_summ.reset_index(inplace=True)
death_veh_summ["CrimeHead"]=death_veh_summ["CrimeHead"].str.rstrip("(Total)")
```

```python
Total_Offenders_Died=new_df["Offenders (Driver/Pedestrian) Died_Total"].sum()
Total_Victims_Died=new_df["Victims Died_Total"].sum()
ax=sns.barplot(x=["Total Offenders Died","Total Victims Died"],y= [Total_Offenders_Died,Total_Victims_Died])
plt.grid(False, axis="x")
ax.bar_label(ax.containers[0])
plt.title("Total Offenders Died vs Total Victims Died in Road Accidents in 2014", fontweight="bold" )
plt.show()
```

#### It seems total number of victims died are quite higher as compared to total number of offenders died in road accidents. Victims death numbers are almost three times of Offenders death numbers

```python
plt.figure(figsize=(20,10))
order=death_veh_summ.nlargest(5,"Offenders (Driver/Pedestrian) Died_Total")["CrimeHead"]
ax=sns.barplot(x="CrimeHead",y="Offenders (Driver/Pedestrian) Died_Total",data=death_veh_summ, order=order)
plt.title("Total Number of Offenders Died for different Crime Heads in 2014", fontweight="bold", fontsize=15)
plt.ylabel("Number of Offenders Died")
ax.bar_label(ax.containers[0])
plt.show()
```

#### Majority of the offenders died when crime head vehicle is Two Wheeler followed by Truck/ Lorry and Car

```python
plt.figure(figsize=(20,10))
order=death_veh_summ.nlargest(5,"Victims Died_Total")["CrimeHead"]
ax=sns.barplot(x="CrimeHead",y="Victims Died_Total",data=death_veh_summ, order=order)
plt.title("Total Number of Victims Died for different Crime Heads in 2014", fontweight="bold", fontsize=15)
plt.ylabel("Number of Victims Died")
ax.bar_label(ax.containers[0])
plt.show()
```

#### Majority of the victims died when crime head vehicle is Two Wheeler and Truck/ Lorry followed by Car and Bus
